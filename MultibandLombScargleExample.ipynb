# Example of a Multi-band Lomb-Scargle Fit

Here's a quick example of doing a multi-band lomb-scargle fit to one of the stripe 82 RR Lyrae stars from Sesar 2010.

First some standard Python imports and settings:

```python
from __future__ import print_function, division

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import seaborn; seaborn.set()
```

## Fetching the data

We'll start by fetching observed light curves from the Sesar 2010 Paper:

```python
from multiband_LS.data import fetch_light_curves
rrlyrae = fetch_light_curves()
lcids = np.fromiter(rrlyrae.ids, dtype=int)
```

Let's plot the first set of light curves

```python
t, y, dy = rrlyrae.get_lightcurve(lcids[0])
```

```python
for i, filt in enumerate('ugriz'):
    plt.errorbar(t[:, i], y[:, i], dy[:, i], fmt='o', label=filt)
plt.legend(ncol=5, loc=0);
```

## Doing a multi-term fit

The ``LombScargleMultiband`` class has the tools needed to compute multiband lomb-scargle powers, and to fit the best model based on these.

Note that we're doing a four-term truncated Fourier fit:

```python
from multiband_LS.lomb_scargle import LombScargleMultiband
model = LombScargleMultiband(Nterms=4)

# Define the filters associated with the data
filts = np.array([['u', 'g', 'r', 'i', 'z']
                  for i in range(t.shape[0])])
```

```python
good = ~np.isnan(t)
model.fit(t[good], y[good], dy[good], filts[good])
```

Let's take a look at lomb-scargle power for this:

```python
periods = np.linspace(0.2, 0.9, 10000)
omegas = 2 * np.pi / periods

P = model.power(omegas)
plt.plot(periods, P);
```

It looks like ~0.62 is the highest power. Let's zoom-in around this:

```python
periods = np.linspace(0.6, 0.63, 10000)
omegas = 2 * np.pi / periods

P = model.power(omegas)
plt.plot(periods, P);
```

Find the optimal period/frequency from this periodogram:

```python
P_best = periods[np.argmax(P)]
omega_best = omegas[np.argmax(P)]
```

Now given this optimal period, we can plot the folded light curves along with the model fit:

```python
tfit = np.linspace(0, P_best, 100)[:, None]
filts = np.array(['u','g','r','i','z'])

yfits = model.predict(tfit, filts, omega_best)

colors = plt.rcParams['axes.color_cycle']

for i, color in enumerate(colors[:5]):
    plt.errorbar(t[:, i] % P_best, y[:, i], dy[:, i], fmt='o', c=color)
    plt.plot(tfit, yfits[:, i], '-', c=color, alpha=0.6, label=filts[i])

plt.gca().invert_yaxis()
plt.legend();
```

From this model, we can plot the trend in color-color space as a function of phase:

```python
plt.errorbar(y[:, 1] - y[:, 2], y[:, 2], yerr=dy[:, 2],
             xerr=np.sqrt(dy[:, 1] ** 2 + dy[:, 2] ** 2), fmt='o')
plt.plot(yfits[:, 1] - yfits[:, 2], yfits[:, 2], '-k', alpha=0.7)
plt.xlabel('g - r'); plt.ylabel('r');
```
