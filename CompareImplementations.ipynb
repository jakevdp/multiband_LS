# Comparing Lomb-Scargle Implementations

We want to make sure here that the astroML versions, which are based on optimized algorithms, match the results of the multiband_LS versions, which are based on less-optimal algorithms.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import seaborn
seaborn.set()
```

```python
from astroML.time_series import lomb_scargle as lomb_scargle_astroML
from astroML.time_series import multiterm_periodogram
from multiband_LS.lomb_scargle import lomb_scargle
```

```python
from astroML.datasets import fetch_LINEAR_sample
data = fetch_LINEAR_sample()
```

```python
lcid = 10022663
t, y, dy = data[lcid].T
```

```python
plt.errorbar(t, y, dy, fmt='o');
```

## Standard Lomb-Scargle

```python
period = np.linspace(0.61, 0.62, 5000)
omega = 2 * np.pi / period

P1 = lomb_scargle_astroML(t, y, dy, omega, generalized=False)
P2 = lomb_scargle(t, y, dy, omega, fit_offset=False)
print("methods match:", np.allclose(P1, P2))

plt.plot(period, P1, label='astroML')
plt.plot(period, P2, label='multiterm')
plt.legend();
```

### Timings

```python
%timeit lomb_scargle_astroML(t, y, dy, omega, generalized=False)
%timeit lomb_scargle(t, y, dy, omega, fit_offset=False)
```

## Generalized Lomb-Scargle

```python
P1 = lomb_scargle_astroML(t, y, dy, omega, generalized=True)
P2 = lomb_scargle(t, y, dy, omega, fit_offset=True)
print("methods match:", np.allclose(P1, P2))

plt.plot(period, P1, label='astroML')
plt.plot(period, P2, label='multiterm')
plt.legend();
```

### Timings

```python
%timeit lomb_scargle_astroML(t, y, dy, omega, generalized=True)
%timeit lomb_scargle(t, y, dy, omega, fit_offset=True)
```

## Multiterm Fits

```python
P1 = multiterm_periodogram(t, y, dy, omega, n_terms=3)
P2 = lomb_scargle(t, y, dy, omega, Nterms=3, fit_offset=True)
print("methods match:", np.allclose(P1, P2))

plt.plot(period, P1, label='astroML')
plt.plot(period, P2, label='multiterm')
plt.legend();
```

### Timings

```python
%timeit multiterm_periodogram(t, y, dy, omega, n_terms=3)
%timeit lomb_scargle(t, y, dy, omega, Nterms=3, fit_offset=True)
```

## Poorly-behaved data

Here we'll check out the difference between standard and generalized Lomb-Scargle...

```python
np.random.seed(0)

lcid = 10022663
t, y, dy = data[lcid].T

# Find best period
period = np.linspace(0.61, 0.62, 5000)
omega = 2 * np.pi / period
P = lomb_scargle_astroML(t, y, dy, omega)
P_best = period[np.argmax(P)]

# Fit a sinusoid
y = 5 * np.mean(dy) * np.sin(2 * np.pi * t / P_best) + np.random.normal(0, dy)
ymean = y.mean()
```

```python
plt.errorbar(t % P_best, y, dy, fmt='o');
```

```python
# Screw-up lomb-scargle by using only the top half of the data

mask = (y > ymean)
t, y, dy = t[mask], y[mask], dy[mask]
plt.errorbar(t % P_best, y, dy, fmt='o');
```

### Standard

```python
period = np.linspace(0.61, 0.62, 5000)
omega = 2 * np.pi / period

P1 = lomb_scargle_astroML(t, y, dy, omega, generalized=False)
P2 = lomb_scargle(t, y, dy, omega, fit_offset=False)
print("methods match:", np.allclose(P1, P2))

print(period[np.argmax(P1)])

plt.plot(period, P1, label='astroML')
plt.plot(period, P2, label='multiterm')
plt.legend();
```

### Generalized

```python
P1 = lomb_scargle_astroML(t, y, dy, omega, generalized=True)
P2 = lomb_scargle(t, y, dy, omega, fit_offset=True)
print("methods match:", np.allclose(P1, P2))

plt.plot(period, P1, label='astroML')
plt.plot(period, P2, label='multiterm')
plt.legend();
```
