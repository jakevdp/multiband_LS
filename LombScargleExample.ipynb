# Lomb-Scargle Example

Here is an example of performing a lomb-scargle periodogram on a LINEAR light curve.

We start with some imports.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import seaborn
seaborn.set()
```

```python
from multiband_LS.lomb_scargle import LombScargle, LombScargleAstroML
```

```python
from astroML.datasets import fetch_LINEAR_sample
data = fetch_LINEAR_sample()
```

```python
lcid = 10022663
t, y, dy = data[lcid].T
```

```python
model1 = LombScargle(fit_offset=True, Nterms=1).fit(t, y, dy)
P_best = model1.find_best_period()
```

```python
period = np.linspace(P_best - 0.005, P_best + 0.005, 5000)
omegas = 2 * np.pi / period

P = model1.power(omegas)
plt.plot(period, P)
plt.xlabel('period (days)'); plt.ylabel('Lomb-Scargle Power');
```

```python
model6 = LombScargle(fit_offset=True, Nterms=6).fit(t, y, dy)
P = model6.power(omegas)

plt.plot(period, P)
plt.xlabel('period (days)'); plt.ylabel('Lomb-Scargle Power');
```

```python
t_fit = np.linspace(0, P_best, 1000)
P_best = period[np.argmax(P)]
omega_best = 2 * np.pi / P_best

plt.errorbar(t % P_best, y, dy, fmt='o')
for model in [model1, model6]:
    y_fit = model.predict(t_fit, omega_best)
    plt.plot(t_fit, y_fit, label="{0} terms".format(model.Nterms))
plt.gca().invert_yaxis()
plt.legend(loc='best');
```

## AstroML Version

For convenience, we've wrapped the astroML version of the lomb-scargle algorithm.
Note that this uses a faster algorithm, but is not as general (i.e. it cannot fit multiple terms).

```python
model_LS = LombScargleAstroML().fit(t, y, dy)
P = model_LS.power(omegas)

plt.plot(period, P)
plt.xlabel('period (days)'); plt.ylabel('Lomb-Scargle Power');
```
